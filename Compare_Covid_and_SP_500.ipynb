```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import plotly.express as px
```

```
sp = pd.read_csv("/content/drive/MyDrive/Data/SP500/SP500.csv")
sp = sp.drop(columns=['Open', 'High', 'Low', 'Close', 'Volume'])
sp.columns = ['Date', 'Close']
sp['Date'] = pd.to_datetime(sp['Date'])
sp['DOY'] = sp['Date'].dt.dayofyear
sp['Year'] = sp['Date'].dt.year
sp.head()
```

```
df1 = sp.loc[sp['Year']== 2020, ['Date', 'Close']]
```

```
df1['Pct'] = (df1['Close'] - df1.iloc[0,1])/df1.iloc[0,1]*100
```

```
df1['Type'] = 'Return'
```

```
df1['DOY'] = df1['Date'].dt.dayofyear
```

```
df1 = df1[['DOY', 'Type', 'Pct']].reset_index(drop=True)
```

```
df1.head()
```

```
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
cases = pd.read_csv(url)

url1 = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
deaths = pd.read_csv(url1)
```

```
cases.head()
```

```
df = cases.loc[cases['Country/Region'].str.contains('US'),'1/22/20':].T
```

```
df = df.reset_index()
```

```
df.columns = ['Date', 'Cases']
```

```
df['Date'] = pd.to_datetime(df['Date'])
```

```
df['Pct'] = (df['Cases'])/df['Cases'].max()*100
```

```
df['DOY'] = df.Date.dt.dayofyear
```

```
df['Type'] = 'Cases'
```

```
df = df[['DOY', 'Type', 'Pct']]
```

```
df.head()
```

```
common = np.intersect1d(df.DOY.values, df1.DOY.values) 
```

```
df3 = pd.concat([df.loc[df['DOY'].isin(common)], df1.loc[df1.DOY.isin(common)]])
```

```
cases.loc[cases['Country/Region'].str.contains('US'),'1/22/20':].T.tail()
```

```
fig = px.bar(df3, x='Type', y='Pct', animation_frame="DOY", range_y=[-50, 100])
fig.show()
```
