Objective: To visualize S&P 500 index using animation.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
import plotly.express as px
```

Prepare the Data for visualization

```
sp = pd.read_csv("/content/drive/MyDrive/Data/SP500/SP500.csv")
```

```
sp = sp.drop(columns=['Open', 'High', 'Low', 'Close', 'Volume'])
```

```
sp.columns = ['Date', 'Close']
```

```
sp['Date'] = pd.to_datetime(sp['Date'])
```

```
sp['DOY'] = sp['Date'].dt.dayofyear
```

```
sp['Year'] = sp['Date'].dt.year
```

```
sp.head()
```

Data Visualization

```
def PlotReturns(year):
  df = sp.loc[sp['Year'].isin(year), ['Year', 'DOY', 'Close']]

  common = np.arange(1,367)

  for y in year:
    start = df.loc[df['Year']==y,'Close'].head(1).values[0]
    df.loc[df['Year']==y,'Return'] = (df.loc[df['Year']==y,'Close'] - start)/start*100

    common = np.intersect1d(common, df.loc[df['Year']==y,'DOY'].values)

  df['Year2'] = df['Year'].apply(lambda x: 'Year ' + str(x))

  fig = px.bar(df.loc[df['DOY'].isin(common),:], 
             x="Year2", y="Return", animation_frame="DOY", range_y=[round(df['Return'].min()*1.2),round(df['Return'].max()*1.2)])
  fig.show()
```

```
PlotReturns([2020])
```

```
def PlotReturnsSeq(year):
  startyear = year[0]
  endyear = year[1]

  yearrange = range(startyear, endyear+1)

  df = sp.loc[sp['Year'].isin(yearrange), ['Year', 'DOY', 'Close']]

  start = df.loc[df['Year']==startyear,'Close'].head(1).values[0]
  
  df['Return'] = (df['Close'] - start)/start*100
  
  df['DOY2'] = df['DOY'] + (df['Year'] - startyear)*365

  fig = px.bar(df.loc[:, ['DOY2', 'Return']], y="Return", 
               animation_frame="DOY2", range_y=[round(df['Return'].min()*1.2),round(df['Return'].max()*1.2)])
  fig.show()
```

```
PlotReturnsSeq([2000, 2002])
```

```
PlotReturnsSeq([2007, 2009])
```
